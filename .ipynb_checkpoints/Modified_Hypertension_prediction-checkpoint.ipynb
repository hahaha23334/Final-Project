```python
# Merge all the datasets
# Integration of the datasets will be done based on the common identifier 'SEQN'
# We will perform an inner join to ensure only records with complete information across the datasets are included

# Merging the datasets
df = pd.merge(dm, diet, on='SEQN', how='inner')
df = pd.merge(df, lab, on='SEQN', how='inner')
df = pd.merge(df, med, on='SEQN', how='inner')
df = pd.merge(df, qs, on='SEQN', how='inner')

# Displaying the shape and the first few rows of the merged dataset
df_shape = df.shape
df_head = df.head()

df_shape, df_head
```

```python
# Data Cleaning and Preprocessing

# Identifying missing values
missing_values = df.isnull().sum()
```

```python
# Dropping columns with a high percentage of missing values
# Setting a threshold, for example, 50% missing values
threshold = 0.5 * len(df)
df_cleaned = df.dropna(thresh=threshold, axis=1)
```

```python
# Handling remaining missing values
# For numerical columns, we'll impute missing values with the median
# For categorical columns, we'll impute missing values with the mode
numerical_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df_cleaned.select_dtypes(include=['object']).columns

for col in numerical_cols:
    df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)

for col in categorical_cols:
    df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)
```

```python
# Encoding categorical variables
# For simplicity, using pandas get_dummies for one-hot encoding
df_encoded = pd.get_dummies(df_cleaned, columns=categorical_cols)

# Displaying the shape of the cleaned and encoded dataset
cleaned_shape = df_encoded.shape

cleaned_shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Exploratory Data Analysis (EDA)

# Example EDA: Analyzing the distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(df_encoded['RIDAGEYR'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age (Years)')
plt.ylabel('Frequency')
plt.show()

# Analyzing the distribution of Blood Pressure (if available in the dataset)
# Assuming 'BPXSY1' represents Systolic Blood Pressure and 'BPXDI1' represents Diastolic Blood Pressure
if 'BPXSY1' in df_encoded.columns and 'BPXDI1' in df_encoded.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df_encoded['BPXSY1'], kde=True, color='blue', label='Systolic BP')
    sns.histplot(df_encoded['BPXDI1'], kde=True, color='red', label='Diastolic BP')
    plt.title('Blood Pressure Distribution')
    plt.xlabel('Blood Pressure (mmHg)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Feature Engineering: Creating Body Mass Index (BMI) from height and weight data
# Assuming 'BMXWT' is weight in kg and 'BMXHT' is height in cm
if 'BMXWT' in df_encoded.columns and 'BMXHT' in df_encoded.columns:
    # Convert height from cm to m and calculate BMI
    df_encoded['BMI'] = df_encoded['BMXWT'] / ((df_encoded['BMXHT'] / 100) ** 2)

    # Analyzing the distribution of BMI
    plt.figure(figsize=(10, 6))
    sns.histplot(df_encoded['BMI'], kde=True)
    plt.title('BMI Distribution')
    plt.xlabel('BMI (kg/m²)')
    plt.ylabel('Frequency')
    plt.show()

# The target variable for hypertension needs to be identified or created
# For instance, we might define hypertension based on blood pressure readings (e.g., BPXSY1 > 140 or BPXDI1 > 90)
# or on medication data indicating hypertension treatment
# This part needs further clarification based on the available data and medical criteria for hypertension diagnosis
```

```python
# Creating the hypertension target variable based on the updated definition

# Checking if SBP and DBP columns exist in the dataset
sbp_col_exists = 'BPXSY1' in df_encoded.columns
dbp_col_exists = 'BPXDI1' in df_encoded.columns

# If SBP and DBP columns exist, use them for hypertension criteria
if sbp_col_exists and dbp_col_exists:
    hypertension_criteria = (df_encoded['BPXSY1'] >= 130) | (df_encoded['BPXDI1'] >= 80) | (df_encoded['BPQ020'] == 1)
else:
    # If SBP and DBP columns do not exist, rely only on medication data
    hypertension_criteria = df_encoded['BPQ020'] == 1

# Creating the hypertension target variable
df_encoded['Hypertension'] = hypertension_criteria.astype(int)

# Analyzing the distribution of the target variable
plt.figure(figsize=(10, 6))
sns.countplot(x='Hypertension', data=df_encoded)
plt.title('Distribution of Hypertension Cases')
plt.xlabel('Hypertension (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Displaying the percentage of hypertensive cases in the dataset
hypertension_rate = df_encoded['Hypertension'].mean() * 100
hypertension_rate
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
import numpy as np

# Selecting relevant features and the target variable
X = df_encoded.drop('Hypertension', axis=1)  # Features
y = df_encoded['Hypertension']  # Target variable

# Splitting the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model 1: Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
```

```python
# Model 2: Logistic Regression
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)
```

```python
# Evaluating the models
rf_accuracy = accuracy_score(y_test, rf_predictions)
lr_accuracy = accuracy_score(y_test, lr_predictions)
```

```python
# Creating confusion matrices for both models
rf_confusion_matrix = confusion_matrix(y_test, rf_predictions)
lr_confusion_matrix = confusion_matrix(y_test, lr_predictions)
```

```python
# Creating classification reports for both models
rf_classification_report = classification_report(y_test, rf_predictions)
lr_classification_report = classification_report(y_test, lr_predictions)

rf_accuracy, lr_accuracy, rf_confusion_matrix, lr_confusion_matrix, rf_classification_report, lr_classification_repor
```
